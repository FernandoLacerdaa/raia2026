# Modelagem

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
vectorizer = TfidfVectorizer()

X = vectorizer.fit_transform(data["comentario"])
```

```python
y = data['sentimento']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
```

# Support vector machines

##### teste size = 0.2

```python
from sklearn.svm import SVC


# Split the dataset into training and testing sets
print("test size é 0.2")
print ("")

acs = []
for i in range (40):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=i)
    clf = SVC(kernel='linear')
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    
    accuracy = accuracy_score(y_test, y_pred)


    acs.append(accuracy)

indices_test = y_test.index


resultado_SVM_02 = pd.DataFrame({
    "comentario": data.loc[indices_test, "comentario"].values,
    "sentimento_predito": y_pred,
    "sentimento_verdadeiro": y_test
})


print (f"STD:{np.array(acs).std()}")
print (f"media:{np.array(acs).mean()}")
```
